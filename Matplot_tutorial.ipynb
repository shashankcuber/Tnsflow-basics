```
#lec -2
#legends titles and labels
x=[1,2,3]
y=[5,7,4]

x2=[1,2,3]
y2=[10,14,12]


plt.plot(x,y,label='First line')
plt.plot(x2,y2,label='Second Line')
plt.xlabel('X axis')
plt.ylabel('Y axis')
#\n for new line
plt.title('Interesting graph\n')

#legends to identify in case of multiplots which graph is which
plt.legend()
plt.show()
```

```
#lec 3
#Bar charts and histogram

x=[2,4,6,8,10]
y=[6,7,8,2,4]

#for multi bars in a graph
x2=[1,2,3]
y2=[10,14,12]

#bar charts
#uncomment to get the bar chart
#plt.bar(x,y,label='First Bar',color='red')
#plt.bar(x2,y2,label='Second Bar',color='c')


#histograms

population_ages=[22,55,62,45,21,22,34,42,78,90,21]
id=[x for x in range(len(population_ages))]
#plt.bar(id,population_ages,label='first')
#bins to identify the range
bins=[0,10,20,30,40,50,60,80,90,100,120,130]
plt.hist(population_ages,bins,histtype='bar',rwidth=0.8)

plt.xlabel('X axis')
plt.ylabel('Y axis')
#\n for new line
plt.title('Interesting graph\n')

#legends to identify in case of multiplots which graph is which
plt.legend()

plt.show()

```

```
#scatter plot lec-4
#used for correlation between diff variables
x=[1,2,3,4,5,6,7,8]
y=[5,2,4,7,6,8,3,7]
#labels for legends in multiplot
#marker changes the point view
# s is for size of the marker
plt.scatter(x,y,label='skitcat',color='k',marker='*',s=25)
plt.legend()# to show the label in graph
plt.show()
```

```
#lec 5 stack plots
days=[1,2,3,4,5]

sleeping=[7,8,6,11,7]
eating=[2,3,4,3,2]
working=[7,8,7,2,2]
playing=[8,5,7,8,13]

#alternate ways to show label as stackplot does not support labels
plt.plot([],[],color='m',label='sleeping',linewidth=5)
plt.plot([],[],color='r',label='eating',linewidth=5)
plt.plot([],[],color='c',label='working',linewidth=5)
plt.plot([],[],color='k',label='eating',linewidth=5)
#stackplot(x,y1,y2,....,colors for each y element)
#in scatter plot we cant have label directly
plt.stackplot(days,sleeping,eating,working,eating,colors=['m','r','c','k'])
plt.legend()
plt.show()
```

```
#lec 6 pie chart

slices=[7,2,2,13]
act=['sleeping','eating','working','playing']
cols=['m','r','c','k']

#shadow just creates a shadow in visualisation
#startangle makes starting act pie from 90 degrees
#explode will take that piece and bulge out
#autopct add percentage to each slice
plt.pie(slices,labels=act,colors=cols,startangle=90,shadow=True,explode=(0,0,1,0),autopct='1.1f%%')
plt.legend()
plt.show()
```

```
#to upload local file to colab
from google.colab import files
uploaded=files.upload()
#lec 16 Live graphs
#importing the animation function

import matplotlib.animation as animation

from matplotlib import style

#style.use('fivethirtyeight')

fig=plt.figure()

ax1=fig.add_subplot(1,2,1)

def animate(i):
  graph_data=open('samplefile.txt','r').read()
  #split it with a new line
  lines=graph_data.split('\n')
  
  xs=[]
  ys=[]
  
  for line in lines:
    if len(line)>1:
      x,y=line.split(',') #just like delimiter
      xs.append(x)
      ys.append(y)
      
      
  ax1.plot(xs,ys)
  
#helps animate based on the function
#animation.FuncAnimation(fig(where we are going to animate our function),animate
#(function we are going to animate),interval=1k i. after evvery 1k msec graph updates)
ani=animation.FuncAnimation(fig,animate,interval=1000)
print(ani)
plt.show()
```

```
#lec 19 Subplots

import random
from matplotlib import style
fig=plt.figure()
style.use('fivethirtyeight')

#to get the basic data to work with
def create_plots():
  xs=[]
  ys=[]

  for i in range(10):
    x=i
    y=random.randrange(10)
    
    
    xs.append(x)
    ys.append(y)
    
  return xs,ys

  
#define an axis
#add_subplot(height,width,plot)
#subplot(2,1,1)
#then this will create two plots like matrices cell of 2x1
ax1=fig.add_subplot(211)
ax2=fig.add_subplot(212)
  
#x,y=create_plots()
  
#ax1.plot(x,y)
  
#x,y=create_plots()

#ax2.plot(x,y)
  
  
a1=fig.add_subplot(221)
a2=fig.add_subplot(222)
a3=fig.add_subplot(212)

x,y=create_plots()
  
a1.plot(x,y)
  
x,y=create_plots()

a2.plot(x,y)

x,y=create_plots()

a3.plot(x,y)

plt.show()

#for subplots in a grid

ax1=fig.subplot2grid((6,1),(0,0),rowspan=2,colspan=1)
x,y=create_plots()
  
ax1.plot(x,y)
```
